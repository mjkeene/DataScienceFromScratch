```python
# Two primary uses for data visualization:
# 1. Explore data
# 2. Communicate data
```

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# add a title
plt.title('Nominal GDP by Year')

# add a label to the x-axis
plt.xlabel('Year')

# add a label to the y-axis
plt.ylabel('Billions of $')

plt.show()
```

```python
# Bar Chart shows how some quantity varies among a 'discrete' set of items

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# can move the bars with this
xs = [i + 0.5 for i, _ in enumerate(movies)]

# can move the x-axis labels with this
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.xlabel('Movie Title')
plt.ylabel('Number of Academy Awards')
plt.title('My Favorite Movies')

plt.bar(xs, num_oscars)
```

```python
# A bar chart is also good for plotting histograms of bucketed numeric values to visualize distribution
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade : grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

# 8 is the bar width, leaves some space between bars
plt.bar([x for x in histogram.keys()], histogram.values(), 8) 
plt.axis([-5, 105, 0, 5]) # x-axis from -5 to 105, y-axis from 0 to 5

plt.xticks([10*i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('Num Students')
plt.title('Distribution of Grades')
plt.show()
```

```python
# The following is a misleading bar chart since the y-axis does not start at 0

mentions = [500, 505]
years = [2013, 2014]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.axis([2012.5, 2014.5, 499, 506])
plt.title('Huge Increase')
```

```python
# Using more sensible axes shows the real comparison

plt.bar(years, mentions, 0.8)
plt.xticks(years)

plt.axis([2012.5, 2014.5, 0, 550])
plt.title('Not Huge Anymore')
```

```python
# Line Charts

variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9) # top center
plt.xlabel('model complexity')
plt.title('Bias-Variance Tradeoff')
```

```python
# Scatterplots visualize the relationship between paired sets of data

friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
```

```python
# Again, the scale of the axes is very important
# Here we have comparable variables with different scales

test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title('Axes are not comparable')
plt.xlabel('test 1 grade')
plt.ylabel('test 2 grade')
```

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.axis('equal') # this makes the axes equal
plt.title('Axes are comparable')
plt.xlabel('test 1 grade')
plt.ylabel('test 2 grade')
```
